# 1. Align dose response, expression, and drug matrices

```python
import pandas as pd
```

```python
# Load GDSC data: RNAseq
expr = pd.read_csv("./gdsc/raw_data/Cell_line_RMA_proc_basalExp.txt", sep="\t", index_col=0)
expr
```

```python
# Remove rows with no information about which gene it is
expr = expr[expr.index.notna()]
expr
```

```python
# Load GDSC data: drug response
dose_response = pd.read_excel("./gdsc/raw_data/GDSC2_fitted_dose_response_27Oct23.xlsx")
dose_response
```

```python
# Change expr column names to match COSMIC IDs
expr.columns = expr.columns.str.replace("DATA.", "")  # Remove 'DATA.' prefix from column names to get COSMIC IDs
expr = expr[
    [col for col in expr.columns if "." not in col]
]  # Some columns have e.g. 905954.1, so are duplicates of e.g. 905954. Remove these columns
expr = expr[expr.columns[1:]]  # Remove first column, as it is not expression data

# Change doseresponse COSMIC_ID to string for matching
dose_response["COSMIC_ID"] = dose_response["COSMIC_ID"].astype(str)
```

```python
common_cell_lines = expr.columns.intersection(
    dose_response["COSMIC_ID"]
)  # Find common cell lines between expression and drug response data
```

```python
dose_response = dose_response[dose_response["COSMIC_ID"].isin(common_cell_lines)]
dose_response
```

```python
expr = expr[common_cell_lines]
expr
```

```python
drugs = pd.read_csv("./gdsc/raw_data/screened_compounds_rel_8.5.csv")
drugs
```

```python
drugs = drugs[drugs["DRUG_ID"].isin(dose_response["DRUG_ID"].unique())]
drugs
```

```python
import pubchempy as pcp

drug_id_to_smiles = {}
for i, j in zip(drugs["DRUG_ID"], drugs["DRUG_NAME"], strict=True):
    compound = pcp.get_compounds(j, "name")
    if compound:
        drug_id_to_smiles[i] = pcp.get_compounds(j, "name")[0].smiles
```

```python
drugs = drugs[drugs["DRUG_ID"].isin(drug_id_to_smiles.keys())]
drugs
```

```python
# Add SMILES to drugs dataframe
drugs["SMILES"] = None
drugs["SMILES"] = drugs["DRUG_ID"].map(drug_id_to_smiles)
drugs
```

```python
dose_response = dose_response[dose_response["DRUG_ID"].isin(drugs["DRUG_ID"])]
dose_response
```

```python
expr = expr[dose_response["COSMIC_ID"].unique()]
expr
```

```python
# Use .pivot() to reshape the dose response df into a matrix
dose_response_matrix = dose_response.pivot(index="DRUG_ID", columns="COSMIC_ID", values="LN_IC50")
dose_response_matrix.notna().sum().sum()
```

```python
# The subset of columns of drugs that we'll actually use
drugs_subset = drugs[["DRUG_ID", "DRUG_NAME", "SMILES"]]
drugs_subset
```

```python
# The subset of columns of dose_response that we'll actually use
dose_response_subset = dose_response[["COSMIC_ID", "TCGA_DESC", "DRUG_ID", "DRUG_NAME", "LN_IC50"]]
dose_response_subset
```

```python
# Now we have a dose response dataframe, an expression dataframe, and a drugs dataframe all aligned to each other
# Save these dataframes
expr.to_csv("./gdsc/processed_data/expr.csv")
dose_response.to_csv("./gdsc/processed_data/dose_response.csv", index=False)
dose_response_subset.to_csv("./gdsc/processed_data/dose_response_subset.csv", index=False)
dose_response_matrix.to_csv("./gdsc/processed_data/dose_response_matrix.csv")
drugs.to_csv("./gdsc/processed_data/drugs.csv", index=False)
drugs_subset.to_csv("./gdsc/processed_data/drugs_subset.csv", index=False)
```

# 3. Create drug embeddings

```python
import torch
from rdkit import Chem
from transformers import AutoModel, AutoTokenizer

# 1. Load Pre-trained Model (Inference Mode)
# "deterministic_eval=True" ensures no dropout/randomness during embedding
model_name = "ibm/MoLFormer-XL-both-10pct"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModel.from_pretrained(model_name, trust_remote_code=True)
model.eval()


def get_molformer_embedding(smiles_list):
    # Canonicalize SMILES first to ensure consistency
    # (The model is sensitive to string variants of the same molecule)
    canon_smiles = [Chem.MolToSmiles(Chem.MolFromSmiles(s)) for s in smiles_list]

    inputs = tokenizer(canon_smiles, padding=True, return_tensors="pt")

    with torch.no_grad():
        outputs = model(**inputs)

    # Use the pooler_output (CLS token equivalent) for the whole-molecule embedding
    # Shape: [batch_size, 768]
    embeddings = outputs.pooler_output
    return embeddings
```

```python
drug_embeddings = {}
for drug_id, drug_smiles in drugs_subset[["DRUG_ID", "SMILES"]].values:
    embedding = get_molformer_embedding([drug_smiles])
    drug_embeddings[str(drug_id)] = embedding.squeeze().numpy()
drug_embeddings
```

```python
drug_embeddings = pd.DataFrame.from_dict(
    drug_embeddings, orient="index", columns=[f"dim_{i}" for i in range(768)]
).rename_axis("Drug_ID")
drug_embeddings
```

```python
drug_embeddings.to_csv("./gdsc/processed_data/drug_embeddings.csv")
```
